# 1월

```python
df1=pd.read_csv("TBDM_TRNSIT_STAT_SUBWAY_202001.csv", encoding="utf-8")
df1=df1[['`MONTH`', '`DAY`', '`STATION_ID`', '`LINE_NM`', '`STATION_NM`', '`GETOFF_CNT`' ]] 
df1.head(5)
```

### 월 / 일 전처리 - day / month

```python
df1["day"]=df1['`DAY`'].astype(str)
df1["day"]=df1["day"].str[1:3]

df1["month"]="01"
# df1["MMDD"]=df1["MMDD"].astype(int)
```

```python
df1 = df1.rename(columns={"`STATION_ID`":"station_ID", 
                         "`LINE_NM`":"line_nm",
                          "`STATION_NM`":"station_nm",
                          "`GETOFF_CNT`":"num",
                         })
df1 = df1.drop(columns=['`MONTH`', '`DAY`'])
```

### 역명 - station_ID

```python
df1["station_ID"]=df1["station_ID"].astype(str)
df1["station_ID"]=df1["station_ID"].str[1:5]
```

### 역이름

```python
df1["station_nm"]=df1["station_nm"].str[1:]
df1["station_nm"]=df1["station_nm"].str[:-1]
```

### 호선명

```python
df1["line_nm"]=df1["line_nm"].str[1:]
df1["line_nm"]=df1["line_nm"].str[:-1]
```

```python
df1.head()
```

### 하차한 사람의 명수

```python
df1["num"]=df1["num"].str[1:]
df1["num"]=df1["num"].str[:-1]
```
