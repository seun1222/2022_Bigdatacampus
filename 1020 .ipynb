## 행정동별_상권변화지표

- 행정동코드, 지역구 합치기

```python
df4 = pd.read_excel("최종_행정동별_상권변화지표.xlsx ")
df4 = df4.drop(columns={"Unnamed: 0","운영영업_개월평균","폐업영업_개월평균"})
df4
```

```python
df4[df4["행정동_코드"]==11740700]
```

```python
df4[df4["행정동_코드"]==11110560]
```

```python
df_4 = df4[["quarter", "행정동_코드", "행정동", "상권_변화_지표"]]
df_4.
```

```python
k=df_4.groupby(["행정동_코드","행정동"], as_index=False)["상권_변화_지표"].agg(",".join)
k # 상권변화지표 
```

```python
k = k.sort_values(["행정동"], ascending=True)
m_df3 = m_df3.sort_values(["행정동"], ascending=True)
```

```python
k[k["행정동"]=="망원1동"] # 관악구 먼저
```
