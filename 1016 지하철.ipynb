```python
sample = pd.read_excel("호선명_역명_2020합계.xlsx")
sample = sample[["line_nm","station_nm","sum_line_station"]]
df=sample.copy()
```

```python
sample2 = pd.read_excel("역명_지역구_최종.xlsx")
sample2 = sample2[["station_nm","지역구"]]
df2=sample2.copy()
```

```python
df2=df2.rename(columns={"station_nm":"station_nm2"})
df2
```

```python
result = pd.concat([df,df2], axis=1, join='inner')
result = result[["line_nm", "station_nm2", "sum_line_station", "지역구"]]
result = result.rename(columns={"station_nm2":"station_nm"})
result
```

```python
result["getoff_sum"]= result.groupby("지역구")["sum_line_station"].transform(sum)
result
```

```python
result2 = result.copy()
result2 = result2[["지역구", "getoff_sum"]].drop_duplicates()
```

```python
sub_goo_sum = result2.sort_values(["getoff_sum"],ascending=False)
sub_goo_sum
```
