```python
import pandas as pd
import numpy as np
```

```python
from sklearn import preprocessing

from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import RobustScaler
```

```python
import warnings
warnings.filterwarnings(action='ignore') # 경고메세지 숨기기
```

```python
fin = pd.read_excel("1101최종본.xlsx")
fin
```

```python
fin.describe()
```

```python
eco_df = fin.copy()
eco_df.set_index("gd_nm",inplace=True)
eco_df = eco_df.drop(columns={"goo_nm","dong_nm"})
```

```python
# standard 정규화를 통해 수요지수 생성

scaler = StandardScaler()
data = scaler.fit_transform(eco_df)

eco_norm = pd.DataFrame(data, index = list(eco_df.index), columns=list(eco_df.columns))
```

```python
eco_norm["cb_rank"] = ""
eco_norm["cg_rank"] = ""
eco_norm["cs_rank"] = ""
eco_norm["ct_rank"] = ""
eco_norm["cr_rank"] = ""
```

```python
n = 69
tu = {9:n*1.00, 8:n*0.96, 7:n*0.89, 6:n*0.77, 5:n*0.60, 4:n*0.40, 3:n*0.23, 2:n*0.11,1:n*0.04}

for i in tu:
    eco_norm = eco_norm.sort_values(["count_bicycle"], ascending = False)
    eco_norm["cb_rank"].iloc[:int(np.floor(tu[i]))] = i

for i in tu:
    eco_norm = eco_norm.sort_values(["count_green"], ascending = False)
    eco_norm["cg_rank"].iloc[:int(np.floor(tu[i]))] = i

for i in tu:
    eco_norm = eco_norm.sort_values(["count_school"], ascending = False)
    eco_norm["cs_rank"].iloc[:int(np.floor(tu[i]))] = i

for i in tu:
    eco_norm = eco_norm.sort_values(["count_top"], ascending = False)
    eco_norm["ct_rank"].iloc[:int(np.floor(tu[i]))] = i

for i in tu:
    eco_norm = eco_norm.sort_values(["ecocar_ratio"], ascending = False)
    eco_norm["cr_rank"].iloc[:int(np.floor(tu[i]))] = i

```

```python
eco_norm
```

```python
eco_norm["sum_rank"] = eco_norm["cb_rank"]+eco_norm["cg_rank"]+eco_norm["cs_rank"]+eco_norm["ct_rank"]+eco_norm["cr_rank"]
eco_norm
```

```python
eco_norm.sort_values(["sum_rank"],ascending=False).head(30)
```

```python
eco_norm.to_excel("fin_dataset_강강송.xlsx")
```
